### Four Russian Algorithm

```python
import itertools
import numpy as np

# Define basic constants
t = 4  # block size
alphabet = ['A', 'G']
substitution_matrix = {
    ('A', 'A'): 1,
    ('G', 'G'): 1,
    ('A', 'G'): -1,
    ('G', 'A'): -1,
}
gap_penalty = -1
```

```python
# Generate all possible 5-letter words from A/G
words = [''.join(p) for p in itertools.product(alphabet, repeat=t)]
```

```python
print(words)
```

```python
def compute_block(top_seq, left_seq, top_deltas, left_deltas):
    """
    Returns deltas of the last row and column after computing a t√ót block
    given top and left sequence segments and their deltas.
    """
    t = len(top_seq)
    dp = np.zeros((t + 1, t + 1), dtype=int)

    # Reconstruct top_scores and left_scores from deltas

    dp[0, 1:] = top_deltas
    dp[1:, 0] = left_deltas

    for i in range(1, t + 1):
        for j in range(1, t + 1):
            match = dp[i-1][j-1] + substitution_matrix[(left_seq[i-1], top_seq[j-1])]
            delete = dp[i-1][j] + gap_penalty
            insert = dp[i][j-1] + gap_penalty
            dp[i][j] = max(match, delete, insert)

    block = dp[1:, 1:]
   

    # Compute deltas for last row and column
    last_row_deltas = [block[-1][0] - left_scores[-1]]
    for j in range(1, t):
        last_row_deltas.append(block[-1][j] - block[-1][j - 1])

    last_col_deltas = [block[0][-1] - top_scores[-1]]
    for i in range(1, t):
        last_col_deltas.append(block[i][-1] - block[i - 1][-1])
    print(last_col_deltas, last_col_deltas)
    return last_row_deltas, last_col_deltas

```

```python
# Precompute all block results
precomputed_blocks = {}

# With gap_penalty = -1, limit deltas to -1 or 0 (no insertion) or 1 (extension from match)
possible_deltas = [-1, 0, 1]

for top_seq in words:
    print("here")

    for left_seq in words:
        for top_deltas in itertools.product(possible_deltas, repeat=t):

            for left_deltas in itertools.product(possible_deltas, repeat=t):
                key = (top_seq, left_seq, tuple(top_deltas), tuple(left_deltas))
                precomputed_blocks[key] = compute_block(top_seq, left_seq, top_deltas, left_deltas)
#

print(f"Precomputed {len(precomputed_blocks)} blocks.")
```

```python
import random

# Step 1: Generate random A/G sequences of length 100
seq1 = ''.join(random.choices(alphabet, k=100))
seq2 = ''.join(random.choices(alphabet, k=100))

# Step 2: Initialize full DP matrix
dp_full = np.zeros((len(seq1) + 1, len(seq2) + 1), dtype=int)

# Initialize first row and column with gap penalties
for i in range(1, len(seq1) + 1):
    dp_full[i][0] = dp_full[i - 1][0] + gap_penalty
for j in range(1, len(seq2) + 1):
    dp_full[0][j] = dp_full[0][j - 1] + gap_penalty

# Step 3: Fill the matrix using 4x4 blocks and precomputed results
for i_block in range(0, len(seq1), t):
    for j_block in range(0, len(seq2), t):
        # Get top sequence (along columns) and left sequence (along rows) for the block
        top_seq = seq2[j_block:j_block + t]
        left_seq = seq1[i_block:i_block + t]

        # Get boundary scores
        top_scores = dp_full[i_block][j_block + 1:j_block + t + 1]
        left_scores = dp_full[i_block + 1:i_block + t + 1, j_block]
        # Compute deltas (differences between consecutive values)
        top_deltas = np.diff(np.insert(top_scores, 0, 0))
        left_deltas = np.diff(np.insert(left_scores, 0, 0))

        key = (top_seq, left_seq, tuple(top_deltas), tuple(left_deltas))
        block = precomputed_blocks.get(key)
        print(block)
        if block is None:
            # Fallback to direct computation (shouldn‚Äôt happen if preprocessing is complete)
            block = compute_block(top_seq, left_seq, top_scores, left_scores)

        # Insert the block into the DP matrix
        for i in range(t):
            for j in range(t):
                if i_block + i + 1 <= len(seq1) and j_block + j + 1 <= len(seq2):
                    dp_full[i_block + i + 1][j_block + j + 1] = block[i][j]

print("DP matrix built using 5x5 Four Russians blocks!")
```
