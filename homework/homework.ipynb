```python
from Bio import SeqIO
from Bio.Seq import Seq

#Task 1

def hamming_distance(s1, s2):
   if (len(s1) != len(s2)):
       return 'Error. Different line lengths.'
   counter = 0
   
   for i in range(len(s1)):
       if s1[i] != s2[i]:
           counter += 1
   return counter

# Данные для проверки
fasta_sequences = list(SeqIO.parse("BioinformaticsCourse2025/homework/1_1/data/f8.fasta", "fasta"))
s1, s2=fasta_sequences[0], fasta_sequences[1]

fasta_sequences2 = list(SeqIO.parse("BioinformaticsCourse2025/homework/1_1/data/gattaca.fasta", "fasta"))
sc1, sc2=fasta_sequences2[0], fasta_sequences2[1]

empty_seq = Seq("")

print(f's1 = {s1.seq}')
print(f's2 = {s2.seq}')
print(f'sc1 = {sc1.seq}')
print(f'sc2 = {sc2.seq}')
```

```python
print('1. Строки разной длины')
print('Результат:', hamming_distance(s1.seq, s2.seq))
```
